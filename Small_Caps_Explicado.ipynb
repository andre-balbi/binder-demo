```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#  Dadod do site fundamentus

file = 'small_caps.csv'

df = pd.read_csv(file, sep='\t')
df.head(5)
```

```python
df = df.apply(lambda x: x.str.replace('.','')) # Deleta os pontos para ficar no padrao brasileiro
df.head()
```

```python
df = df.apply(lambda x: x.str.replace('%','')) # Deleta o sinal de '%'
df.head()
```

```python
df = df.apply(lambda x: x.str.replace(',','.')) # Substitui os a virgula por pontos para ficar no padrao americano
df.head()
```

```python
df.set_index('Papel', inplace=True) # Definir coluna papel como index
df.head()
```

```python
df = df.astype(float) # Transforar os itens str para float
df.head()
```

```python
max_liq = 20000000 # Maxima liquidez
min_liq = 100000 # Minima liquidez

mask = (df['Liq.2meses'] <= max_liq) & (df['Liq.2meses'] >= min_liq)
```

```python
df = df[mask]
df.info()
```

### ***FORMULA MAGICA DE GREENBLATT***


EV/EBIT --> Preco

ROIC    --> Rentabilidade

Ranking de ROICE   --> Pontuar

Ranking de EV/EBIT --> Pontuar

Soma --> Selecionar as maiores pontuacoes

```python
df.head()
```

```python
df = df[df['ROIC'] > 0] # Selecionando apenas os ROIC positivo
df.head()
```

```python
df = df[['ROIC', 'EV/EBITDA']]
df.head()
```

```python
# Classificar do MENOR para o MAIOR (quanto MAIOR o ROIC, maior a pontuacao)
df = df.sort_values(by='ROIC').reset_index().reset_index().set_index('Papel').rename(columns={'index': 'Rank_ROIC'}) 
df
```

```python
# Classificar do MAIOR para o MENOR (quanto MENOR o EV/EBITDA, maior a pontuacao)
df = df.sort_values(by='EV/EBITDA', ascending=False).reset_index().reset_index().set_index('Papel').rename(columns={'index': 'Rank_EV/EBITDA'})
df
```

```python
df['Soma_Rank'] = df['Rank_EV/EBITDA'] + df['Rank_ROIC'] # Somando a pontuacao 
df = df[['ROIC', 'EV/EBITDA', 'Rank_ROIC', 'Rank_EV/EBITDA', 'Soma_Rank']].sort_values(by='Soma_Rank', ascending=False) # Organizando as colunas
df.head(10) # As 10 melhores pontuadas
```
